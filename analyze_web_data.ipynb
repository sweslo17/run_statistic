```python
%matplotlib inline
import sklearn
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import sklearn
import requests
import datetime
import pickle
```

```python
def analyze_df(df,axs):
    df['OfficialTime'].hist(ax=axs, bins=20)
    print("mean: {}, median:{}, diff:{}".format(df['seconds'].mean(),df['seconds'].median(),df['seconds'].mean() - df['seconds'].median()))
```

```python
fig, axs = plt.subplots(2,3,figsize=(25,15))
count = 0
for i in range(2011,2017):
    df = pickle.load(open("data/tpe/{}.pkl".format(i), 'rb'))
    analyze_df(df, axs[int(count/3)][count%3])
    count += 1
```
